```python
# calculated column
sepalarea= iris_without_species['sepal_length'] * iris_without_species['sepal_width']
sepalarea.head()
```

```python
iris_without_species['sepal_area'] = sepalarea

iris_without_species.head()
```

```python
# petal area

arief = iris_without_species['petal_length'] * iris_without_species['petal_width']
arief.head()
```

```python
iris_without_species['petal_area'] = arief
iris_without_species.head()
```

```python
# filter out parts of Dataframes using pandas
# only part of iris_without_species value of sepal_area is > 15

largeflower = iris_without_species[iris_without_species['sepal_area']> 15] 

largeflower.head(10)
```

```python
largeflower = iris_without_species[iris_without_species['sepal_area']> 18] 

largeflower.head(10)
```
