```python
# data

# bar graph 

# create a list of students
# grades

students=["nur", "siha","suha","shazatul","asy","anis"]
grades = [58,95,95,76,60,98]
```

```python
pip install matplotlib 
```

```python
import matplotlib.pyplot as plt

#create a bar plot

plt.bar(students,grades) #specify propeties of our bar chart

plt.title("Chart of students grades")

plt.xlabel('Students Name')

plt.ylabel('Marks')

plt.show()
```

```python
plt.bar(students,grades,color = 'purple') #cara nak tukar color bar chart

plt.title("Chart of students grades")

plt.xlabel('Students Name')

plt.ylabel('Marks')

plt.show()
```

```python
plt.scatter(students,grades) #specify propeties of our bar chart

plt.title("Chart of students grades")

plt.xlabel('Students Name')

plt.ylabel('Marks')

plt.show()
```

```python
plt.plot(students,grades) 

plt.title("Chart of students grades")

plt.xlabel('Students Name')

plt.ylabel('Marks')

plt.show()
```

```python
# all of them in the same place

# create a figure of width, height and position - row size, and columns size
plt.figure(figsize = (15,5))   #  width, height and position - define the size  

# sub plot 

plt.subplot(1,3,1) # rows, columns position 
plt.bar(students, grades, color = 'green') # specify properties of our bar chart


# sub plot 

plt.subplot(1,3,2) # rows, columns second
plt.plot(students, grades, color = 'green') # specify properties of our bar chart


# sub plot 

plt.subplot(1,3,3) # rows, columns third
plt.scatter(students, grades, color = 'green') # specify properties of our bar chart
```
